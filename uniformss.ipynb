```python
sng=raw_input()
nn=int(raw_input())
x=[]

for ix in range(nn):
    x.append(int(raw_input()))
w={}    
for ix in range(ord('a'),ord('a')+26):
    w[chr(ix)]=ord(chr(ix))-ord('a')+1
    
    
def weight(string):
    s=0
    for ix in range(len(string)):
        
        s=s+w[string[ix]]
        
        
    return s



def uniformstrings(string):
    
    ustrings=[]
    
    
    string=list(string)
    
#    print string,len(string)
    ix=0
    while ix<len(string):
        
        pos=ix
        
        while string[pos]==string[ix]:
            
            ix=ix+1
            #print ix
            if ix==len(string):
                break

            
        end=ix-1
#        print pos,end
                    
        
        ustrings.append(string[pos:end+1])
        
        
        if ix==len(string):
            break
            
    return ustrings
            
    
us= uniformstrings(sng)


def extractweights(l):
    
    wt=[]
    for ix in range(len(us)):
        s=''
        
        for iy in range(len(l[ix])):
        
            s=s+l[ix][iy]
        
            wt.append(weight(s))
        
        
        
    return wt


    
finalw= extractweights(us)


for ix in range(nn):
    
    flag=0
    
    for iy in range(len(finalw)):
        
        if x[ix]==finalw[iy]:
            flag=1
            
    if flag==1:
        
        print'Yes'
        
    else:
        print'No'
```
